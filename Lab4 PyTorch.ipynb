# Lab4 Pytorch

```python
import numpy as np
import torch
```

```python
array = [[1,2,3],[4,5,6]]  #初始化数据
```

```python
np_array = np.array(array)
print("Numpy Array Type: {}".format(type(np_array)))
print("Numpy Array Shape:{}".format(np_array.shape))
print(np_array)
```

```python
tensor = torch.Tensor(array)
```

```python
print("PyTorch Array Type:",tensor.type)
print("PyTorch Array Shape:",tensor.shape)
print(tensor)
```

&emsp;&emsp;这是因为PyTorch张量默认是<u>浮点型</u>的。
因此，如果您输入整数列表，则将它们转换为浮点数，
并在数字后面添加一个小数点以表示其为浮点数。

 ## Tensors的基本运算

```python
a = torch.rand(3,4)
b = torch.ones(3,4)
print(a)
print(b)
```

```python
print(a.reshape(2,6))  #重组成2*6的矩阵
print(a + b)
print(a - b)
print(a * b)
print(b / a)
print(a @ b.T)
print((b.T @ a).mean())  #求均值
print((b.T @ a).std())   #求标准差
print(a.mean())
print(a.mean().item())  #将tensor转换为标量
print(type(a.mean().item()))
```
